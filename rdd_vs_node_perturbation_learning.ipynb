## Node perturbation vs RDD learning

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)

#%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rand
import pandas as pd
import seaborn as sns
from matplotlib import animation, rc
from IPython.display import HTML

from lib.lif import LIF, ParamsLIF
```

```python
q = 3                      #Dimension of learnt vector
dt = 0.001                 #Simulation timestep
DeltaT = 50               #Number of timebins over which learning rule is applied
tsim = 500                #Total simulation time
T = int((tsim/dt)/DeltaT)  #Number of learning blocks
Nsims = 4                  #Number of repeated simulations
c = 0.5                   #Correlation coefficient
x_input = 0                #Input
n = 2                      #Number of neurons
sigma = 10                 #Their noise level
sigma_perturb = 20        #Perturbation noise level
mu = 1                     #Spiking threshold
tau = 1                    #Neuron timescale
eta = 1e-2                    #Cost gradient learning rate (RDD)
#With linear correction
etas = np.array([1e-2, 2e-3, 2e-3])
#Without linear correction
#etas = np.array([1e-2, 0, 0])

epsilon = 2e5              #Weight learning rate (RDD)
epsilon_np = 1e-1          #Weight learning rate (NP)
epsilon_np_baseline = 1e-1 #Weight learning rate with baseline (NP)
p = 0.5                    #Learning window
tau_s = 0.20               #Output filter timescale

N = 19
wmax = 20
wmin = 2

nsims = 1
fn_in = './sweeps/param_w_N_%d_nsims_%d_c_%f_deltaT_simulations.npz'%(N, nsims, c)
sims = np.load(fn_in)
hs = sims['hs']
vs = sims['vs']
wvals = sims['wvals']

t_filter = np.linspace(0, 1, 2000)
exp_filter = np.exp(-t_filter/tau_s)
exp_filter = exp_filter/np.sum(exp_filter)
ds = exp_filter[0]

B1 = 1
B2 = 2
x = .01
y = 0.1
z = 0
cost_fun = lambda s1, s2: (B1*s1-x)**2 + (z+B2*s2 - B2*(B1*s1-y)**2)**2

w1 = 2
w2 = 17

W_rdd = np.array([w1,w2], 'float64')
W_np = W_rdd.copy()
W_np_baseline = W_np.copy()
```

```python
#Setup neurons
params_lif = ParamsLIF(sigma = sigma, tau = tau, mu = mu, c = c)
lif_rdd = LIF(params_lif, t = DeltaT*dt, t_total = tsim)
lif_rdd.W = W_rdd.copy()
lif_rdd.x = x_input

lif_np = LIF(params_lif, t = DeltaT*dt, t_total = tsim)
lif_np.W = W_np.copy()
lif_np.x = x_input

lif_npb = LIF(params_lif, t = DeltaT*dt, t_total = tsim)
lif_npb.W = W_np_baseline.copy()
lif_npb.x = x_input
```

## Learning with RDD

```python
W_np_trace = np.zeros((n, T, Nsims))
W_np_baseline_trace = np.zeros((n, T, Nsims))
W_rdd_trace = np.zeros((n, T, Nsims))
V_trace = np.zeros((n, T, Nsims))

def convolve_online(s, h, kernel, t_offset):
    #print np.nonzero(h)[0]
    for idx in np.nonzero(h)[0]:
        st = t_offset + idx
        en = min(s.shape[0], st + kernel.shape[0])
        ln = en-st
        #print st,en,ln,idx,t_offset
        s[st:en] += kernel[0:ln]
    
for idx in range(Nsims):
    #Create the filtered output vectors now, fill them as we go
    s1 = np.zeros(int(T/dt))
    s2 = np.zeros(int(T/dt))

    s1_np = np.zeros(int(T/dt))
    s2_np = np.zeros(int(T/dt))
    
    s1_npb = np.zeros(int(T/dt))
    s2_npb = np.zeros(int(T/dt))
    s1_npb2 = np.zeros(int(T/dt))
    s2_npb2 = np.zeros(int(T/dt))

    
    lif_rdd = LIF(params_lif, t = DeltaT*dt, t_total = tsim)
    lif_rdd.W = W_rdd.copy()
    lif_rdd.x = x_input

    lif_np = LIF(params_lif, t = DeltaT*dt, t_total = tsim)
    lif_np.W = W_np.copy()
    lif_np.x = x_input

    lif_npb = LIF(params_lif, t = DeltaT*dt, t_total = tsim)
    lif_npb.W = W_np_baseline.copy()
    lif_npb.x = x_input

    count = np.zeros(n)
    print("N=%d"%idx)

    lif_rdd.count = 0
    bt = [False, False]

    V = np.zeros((n, q))
    #Then just repeat the learning rule as before
    dV = np.zeros(V.shape)

    for j in range(T):
        #Simulate LIF for RDD
        #print("t = %d"%j)
        (v_raw, h_raw, _, _, u_raw) = lif_rdd.simulate(DeltaT)

        t_offset = j*DeltaT
        convolve_online(s1, h_raw[0,:], exp_filter, t_offset)
        convolve_online(s2, h_raw[1,:], exp_filter, t_offset)
        cost = cost_fun(s1[t_offset+DeltaT], s2[t_offset+DeltaT])
        nB = h_raw.shape[1]/DeltaT 
        um = u_raw.reshape((n, nB, DeltaT))
        u = np.max(um, 2)
        
        for k in range(n):
            if (u[k,0] > mu - p) & (u[k,0] < mu):
                if bt[k] == False:
                    ahat = np.array([1, 0, -(u[k,0]-mu)])
                    ahat2 = np.array([1, 0, -(u[k,0]-mu)])
                    #ahat = np.array([1, 0, 0])
                    dV[k,:] += (np.dot(V[k,:], ahat)+cost)*ahat2                    
                    bt[k] = True
            elif (u[k,0] < mu + p) & (u[k,0] >= mu):
                #Only do the update when firing...
                if bt[k] == True:
                    ahat = np.array([1, -(u[k,0]-mu), 0])
                    ahat2 = np.array([1, -(u[k,0]-mu), 0])
                    #ahat = np.array([1, 0, 0])
                    dV[k,:] += (np.dot(V[k,:], ahat)-cost)*ahat2                                        
                    count[k] += 1
                    #V[k,:] = V[k,:] - eta*dV[k,:]#/(count[k]+1)
                    V[k,:] = V[k,:] - np.dot(etas,dV[k,:])#/(count[k]+1)
                    dV[k,:] = np.zeros((1,q))
                    bt[k] = False
                
        #At end of episode, update weights according to V for RDD
        lif_rdd.W -= epsilon*np.multiply(V[:,0], mu_mean(lif_rdd.W))#/(j+100)
        lif_rdd.W = np.maximum(np.minimum(lif_rdd.W, wmax), wmin)
        W_rdd_trace[:,j,idx] = lif_rdd.W
        V_trace[:,j,idx] = V[:,0]
        
        #print V
    
        #Node perturbation:
        #Simulate LIF for NP without baseline
        (v_raw, h_raw, _, _, eligibility) = lif_np.simulate_perturbed(sigma_perturb)
        convolve_online(s1_np, h_raw[0,:], exp_filter, t_offset)
        convolve_online(s2_np, h_raw[1,:], exp_filter, t_offset)
        #cost_perturbed = cost_fun(s1_np[t_offset+DeltaT], s2_np[t_offset+DeltaT])
        cost_perturbed = np.mean(cost_fun(s1_np[t_offset:t_offset+DeltaT], s2_np[t_offset:t_offset+DeltaT]))
        lif_np.W -= epsilon_np*cost_perturbed*eligibility
        lif_np.W = np.maximum(np.minimum(lif_np.W, wmax), wmin)
        W_np_trace[:,j,idx] = lif_np.W
        #W_np_trace[:,j,idx] = epsilon_np*cost_perturbed*eligibility
        
        (_, h_raw, _, _, eligibility) = lif_npb.simulate_perturbed(sigma_perturb)
        convolve_online(s1_npb, h_raw[0,:], exp_filter, t_offset)
        convolve_online(s2_npb, h_raw[1,:], exp_filter, t_offset)
        #cost_perturbed = cost_fun(s1_npb[t_offset+DeltaT], s2_npb[t_offset+DeltaT])
        cost_perturbed = np.mean(cost_fun(s1_npb[t_offset:t_offset+DeltaT], s2_npb[t_offset:t_offset+DeltaT]))
        (_, h_raw, _, _, _) = lif_npb.simulate_perturbed(0)
        convolve_online(s1_npb2, h_raw[0,:], exp_filter, t_offset)
        convolve_online(s2_npb2, h_raw[1,:], exp_filter, t_offset)
        #cost = cost_fun(s1_npb2[t_offset+DeltaT], s2_npb2[t_offset+DeltaT])
        cost = np.mean(cost_fun(s1_npb2[t_offset:t_offset+DeltaT], s2_npb2[t_offset:t_offset+DeltaT]))

        lif_npb.W -= epsilon_np*(cost_perturbed-cost)*eligibility
        lif_npb.W = np.maximum(np.minimum(lif_npb.W, wmax), wmin)
        W_np_baseline_trace[:,j,idx] = lif_npb.W
        #W_np_baseline_trace[:,j,idx] = epsilon_np*(cost_perturbed-cost)*eligibility

        #print j, eligibility, cost_perturbed, cost, cost_perturbed - cost

        
print count
```
