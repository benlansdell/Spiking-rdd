```python
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rand
import pandas as pd

from pyESN import ESN
```

```python
esn = ESN(1, 1, n_reservoir = 400, sparsity = 0.95)
```

```python
a = esn.predict(3*np.ones(100))
```

```python
a.shape
```

```python
b=plt.imshow(a)
plt.colorbar(b)
```

```python
for x in range(10):
    a = esn.predict((x-5)*np.ones(100))
    fig,ax = plt.subplots(1,1)
    b=ax.imshow(a)
    plt.colorbar(b)
```

See how this guy does learning a simple quadratic function

```python
#            inputs: array of dimensions (N_training_samples x n_inputs)
#            outputs: array of dimension (N_training_samples x n_outputs)
#            inspect: show a visualisation of the collected reservoir states
n = 20000
eps = np.random.randn(n,1)*10
inputs = np.zeros(n)
for i in range(1,n):
    inputs[i] = 0.95*inputs[i-1]+eps[i]
outputs = inputs**2
plt.plot(inputs)
plt.plot(outputs)
esn.fit(inputs, outputs, False)
```

```python
inputs = np.random.randn(1000,1)*10
ypred = esn.predict(inputs)
plt.plot(inputs, inputs**2, '.r')
plt.plot(inputs, (ypred), '.')
#plt.ylim([-1000,2000])
```
