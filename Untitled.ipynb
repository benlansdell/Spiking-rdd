```python
import cPickle, gzip
import numpy as np
import random
# Load the dataset
f = gzip.open('mnist.pkl.gz', 'rb')
train_set, valid_set, test_set = cPickle.load(f)
f.close()
```

1. randomly choose an MNIST image from MNIST training data.
2. simulate network for D milliseconds.
   a) Over this interval, at each time bin, for each of the 784 input neurons, sample whether a neuron spikes or not at a rate proportional to the pixel intensity
   b) Run the rest of the simulation with this input vector
3. goto 1.

```python
rand_num=random.randint(0,train_set[0].shape[0])
mnist_rand=(train_set[0][rand_num]) 
label_rand=train_set[1][rand_num]
```

```python
lamba=0.02
#generate random 1 X M matrix (this will be a spike train)
D=1000

new_var=np.zeros([784,D])
for i in range(mnist_rand.shape[0]):
    x = np.random.uniform(0,1,D)
    #spike if pixel intensity is grater than 1-lamba*I_i 
    store_bool = x < (lamba*mnist_rand[i])
    new_var[i,:] = store_bool*1
```
